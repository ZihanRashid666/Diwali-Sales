```python
df2_reverse=df2[::-1]
```

```python
customer_genderdistribution=data[["Cust_name","Gender"]]
customer_genderdistribution
```

```python
customer_genderdistributions=customer_genderdistribution["Gender"].value_counts()
customer_genderdistributions
```

```python
customer_genderdistributions.name="Number of people"
customer_genderdistributions=customer_genderdistributions.reset_index()
customer_genderdistributions
```

```python
ax=plt.bar(customer_genderdistributions["Gender"],customer_genderdistributions["Number of people"])

for bar in ax:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2.0, height, f'{height}', ha='center', va='bottom')

plt.xlabel('Gender')
plt.ylabel('Number of People')
plt.title('Gender Distribution of Customers')
```

# people falling into each age group and gender combination

```python
#how many people fall into each age group and gender combination
customer_agegroups=data[['Age Group', 'Gender']].value_counts().reset_index(name='Number of people')
customer_agegroups
```

```python
#In the provided data, the age group 26-35 has 3271 females and 1272 males, 
#making it the age group with the highest number of people

ax=sns.barplot(x = 'Age Group',y= 'Number of people',hue="Gender" ,data = customer_agegroups)

for container in ax.containers:
    ax.bar_label(container)
```

```python
customer_stategroups=data[['Age Group', 'State']].value_counts().reset_index(name='Number of people in state')

top_10_customer_stategroups = customer_stategroups.sort_values(by='Number of people in state', ascending=False)[:10]

top_10_customer_stategroups
```

```python
sns.barplot(x="Age Group",y="Number of people in state",hue="State",data=top_10_customer_stategroups)

for container in ax.containers:
    ax.bar_label(container)

plt.title("Number of people in age group from state")
```
